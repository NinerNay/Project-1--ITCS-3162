## Introduction


In today’s society, social media is everywhere. Whether it’s scrolling through TikTok, uploading stories on Instagram, or messaging friends through Snapchat, these platforms have become second nature in our daily routines. We often rely on these platforms to stay connected with close family and friends, especially when our busy schedules and academics become too much to handle. But this raises the question: how connected are we truly to these relationships?

This project explores whether social media strengthens our relationships, or are these relationships slowly diminishing before our own eyes without knowing.We can do so by answering these questions:
* How does social media vary across age groups?
* Does higher social media usage directly correlate to conflicts in relationships?
* Are students in relationships more or less likely to have high social media score?


## Introducing the Data

The dataset for this project was obtained from kaggle. It contains survey responses from students between the ages of 16 and 25, collected in the first quarter of 2025. The survey is designed to measure how social media correlates with relationships, academics, and mental health. This dataset uses validated psychological scales, such as the Bergen Social Media Addiction Scale(BSMAS), to measure levels of addiction, specifically pertaining to social media. The Bergen Social Media Addiction is a 9-question measure of people's addictive tendency with the focus being social media. This is self report tool with a range of 1-5, wiht 1 being "very rarely" and 5 being "very often". Furthermore, the total score ranges from 6-30, with the highest showing signs of problematic social media use.
